## Distance Metrics used in Machine Learning

### Hemant Thapa

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

### Manhattan Distance

```python
#declaring function for various distance 
def manhattan_distance(x, y):
    return np.sum(np.abs(x-y))
```

### Euclidean distance 

```python
def euclidean_distance(x, y):
    return np.sqrt(np.sum((x-y)**2))
```

### Minkowski Distance

```python
# p=3 for general Minkowski
def minkowski_distance(x, y, p=3):
    return np.sum(np.abs(x-y)**p)**(1/p)
```

```python
#x1, y1
point1 = np.array([0, 0])
#x2, y2
point2 = np.array([3, 4])
```

```python
#calculating distances
manhattan_dist = manhattan_distance(point1, point2)
euclidean_dist = euclidean_distance(point1, point2)
minkowski_dist = minkowski_distance(point1, point2, p=3)
```

```python
print(f"Manhattan Distance: {manhattan_dist} units")
print(f"Euclidean Distance: {euclidean_dist} units")
print(f"Minkowski Distance (p=3): {minkowski_dist} units")
```

```python
fig, ax = plt.subplots()
points = np.array([point1, point2])
ax.plot(points[:,0], points[:,1], 'o')
#annotations
for i, txt in enumerate(['Point1 (0,0)', 'Point2 (3,4)']):
    ax.annotate(txt, (points[i,0], points[i,1]))
plt.grid(True)
#plotting the distances 
plt.plot([point1[0], point2[0]], [point1[1], point1[1]], 'r--', label='Manhattan (Horizontal)')
plt.plot([point2[0], point2[0]], [point1[1], point2[1]], 'r--')
plt.plot(points[:,0], points[:,1], 'b-', label='Euclidean')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Distance Comparisons')
plt.grid(True, ls='--', alpha=0.5)
plt.show()
```

### Scipy for calcualting distance 

```python
A = np.array([0, 0])
B = np.array([3, 4])
```

```python
#distances using scipy.spatial.distance
manhattan_dist_scipy = distance.cityblock(A, B)
euclidean_dist_scipy = distance.euclidean(A, B)
#equivalent to Manhattan
minkowski_dist_scipy_p1 = distance.minkowski(A, B, p=1)  
#equivalent to Euclidean
minkowski_dist_scipy_p2 = distance.minkowski(A, B, p=2)  
minkowski_dist_scipy_p3 = distance.minkowski(A, B, p=3)
```

```python
print(f"Manhattan Distance (scipy): {manhattan_dist_scipy} units")
print(f"Euclidean Distance (scipy): {euclidean_dist_scipy} units")
print(f"Minkowski Distance (p=1, scipy): {minkowski_dist_scipy_p1} units")
print(f"Minkowski Distance (p=2, scipy): {minkowski_dist_scipy_p2} units")
print(f"Minkowski Distance (p=3, scipy): {minkowski_dist_scipy_p3} units")
```
